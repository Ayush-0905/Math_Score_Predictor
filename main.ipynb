```python
import pandas as pd 

df = pd.read_csv('data.csv')

df.head()
```

```python
df['gender']=df['gender'].apply(lambda x: 1 if x == 'male' else 0)
df['race/ethnicity'].value_counts()
df['gender'].value_counts()#df.info()
```

```python
df['gender'].astype('int32')
df.info()
```

```python
df.drop(columns=['race/ethnicity'],inplace=True)
```

```python
df.info()
df['parental level of education'].value_counts()
```

```python
def datap(x):
    if x=='some college':
        return 1
    elif x=='high school':
        return 2
    elif x=="associate's degree":
        return 3
    elif x=="bachelor's degree":
        return 4
    elif x=="master's degree":
        return 5
    elif x=="some high school":
        return 6
df['parental level of education']=df['parental level of education'].apply(datap)
```

```python
df['parental level of education'].value_counts()
df.info()
df['lunch'].value_counts()
```

```python
df['parental level of education'].value_counts()
```

```python
df['lunch']=df['lunch'].apply(lambda x:1 if x=="standard" else 0)
df.info()
```

```python
df['test preparation course'].value_counts()
df['test preparation course']=df['test preparation course'].apply(lambda x:1 if x=='completed' else 0)
df.info()
```

```python
df.corr()['math score']
x=df.drop(columns=['math score'])
y=df['math score']
```

```python
sh=x.shape
sh
```

# Linear Regression using Machine Learning

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)

y_pred = lr.predict(x_test)

from sklearn.metrics import r2_score

r2_score(y_test,y_pred)

```

# Deep Learning

```python
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation

Model = Sequential()

Model.add(Dense(64, input_dim=6,activation='linear'))
Model.add(Dense(64, activation='linear'))
Model.add(Dense(1, activation='linear'))

Model.compile(loss='mse', optimizer='adam', metrics=['mae'])
```

```python
Model.fit(x_train, y_train,epochs=999)
```

```python
Model.save("model.keras")
m=Model.predict(x_test)
r2_score(y_test,m)
```
